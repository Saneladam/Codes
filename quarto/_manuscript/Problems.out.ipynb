## Introduction

We solve the 1D Poisson equation:

$$
-\\frac{d^2 u}{dx^2} = f(x)
$$

````
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 100)
plt.plot(x, np.sin(np.pi*x))
plt.xlabel("x")
plt.ylabel("u(x)")
plt.show()
````
